```python
import pandas as pd

df = pd.read_csv('master_dataset.csv')
print(df.shape)
print(df.columns.tolist())
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df = df[df['country_iso'].notna()].copy()
print(df.shape)
```

```python
print(df[['total_displaced', 'gdp_per_capita', 'unemployment_rate', 'inflation_cpi']].isnull().sum())
```

```python
small_states = ['Anguilla', 'Andorra', 'Brunei Darussalam', 'Cook Islands', 
                'Finland', 'Micronesia (Federated States of)', 'Iceland', 
                'Kiribati', 'Luxembourg', 'Monaco', 'Marshall Islands', 
                'Malta', 'Montserrat', 'Niue', 'Norway', 'Nauru', 'Palau', 
                'San Marino', 'Turks and Caicos Islands', 'Tuvalu', 
                'British Virgin Islands', 'Vanuatu', 'Samoa']

df.loc[df['country_name'].isin(small_states), 'total_displaced'] = \
    df.loc[df['country_name'].isin(small_states), 'total_displaced'].fillna(0)

print(df['total_displaced'].isnull().sum())
```

```python
print(df[df['total_displaced'].isnull()]['country_name'].unique())
```

```python
df = df[df['country_name'] != 'World'].copy()
print(df.shape)
print(df['total_displaced'].isnull().sum())
```

```python
print(df[df['gdp_per_capita'].isnull()]['country_name'].unique())
```

```python
print(df[df['country_name'].isin(['Cuba', 'Eritrea', "Democratic People's Republic of Korea", 'Yemen', 'South Sudan'])]\
    [['country_name', 'year', 'gdp_per_capita']].groupby('country_name')['gdp_per_capita'].apply(lambda x: x.isnull().sum()))
```

```python
df['gdp_per_capita'] = df.groupby('country_name')['gdp_per_capita'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

print(df['gdp_per_capita'].isnull().sum())
```

```python
df['unemployment_rate'] = df.groupby('country_name')['unemployment_rate'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

df['inflation_cpi'] = df.groupby('country_name')['inflation_cpi'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

print(df[['unemployment_rate', 'inflation_cpi']].isnull().sum())
```

```python
df.to_csv('master_dataset_clean.csv', index=False)
print('Salvato! Shape finale:', df.shape)
```

```python
import numpy as np
df['log_total_displaced'] = np.log1p(df['total_displaced'])
df.to_csv('master_dataset_clean.csv', index=False, float_format='%.2f', decimal='.')
print('Salvato!')
```

```python
top30 = df.groupby('country_iso')['total_displaced'].max().nlargest(30).index
df_top30 = df[df['country_iso'].isin(top30)].copy()
df_top30.to_csv('top30_displaced.csv', index=False, float_format='%.2f', decimal='.')
print(df_top30['country_name'].unique())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['displacement_norm'] = scaler.fit_transform(df[['total_displaced']])
df['mortality_norm'] = scaler.fit_transform(df[['mortality_5t14']])
df['crisis_index'] = (df['displacement_norm'] + df['mortality_norm']) / 2

print(df[df['country_name']=='Ukraine'][['year','total_displaced','mortality_5t14','crisis_index']].tail(6))
```

```python
df.to_csv('master_dataset_clean.csv', index=False, float_format='%.2f', decimal='.')

df_recent = df[(df['year'] >= 2019) & (df['is_top30'] == True)].copy()
df_recent_avg = df_recent.groupby('country_name').agg(
    avg_displaced=('total_displaced','mean'),
    avg_mortality=('mortality_5t14','mean'),
    avg_crisis_index=('crisis_index','mean'),
    region=('region','first')
).reset_index().sort_values('avg_crisis_index', ascending=False)

print(df_recent_avg.head(10))
```

```python
df_recent_avg.to_csv('crisis_2019_2022.csv', index=False, float_format='%.2f', decimal='.')
print('Salvato!')
```
